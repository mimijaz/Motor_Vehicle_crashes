### Clean the Data

```python
# Drop duplicates if any
df.drop_duplicates(inplace=True)

# Fill missing values (example: fill missing boroughs with 'Unknown')
df['BOROUGH'].fillna('Unknown', inplace=True)

# Convert date column to datetime
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'], format='mixed')

# Display updated info
df.info()
```

### Accidents by Borough

```python
# Count number of crashes per borough
borough_counts = df['BOROUGH'].value_counts()

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=borough_counts.index, y=borough_counts.values, palette='coolwarm')
plt.xlabel("Borough")
plt.ylabel("Number of Accidents")
plt.title("Number of Accidents by Borough")
plt.xticks(rotation=45)
plt.show()
```

### Accidents Over Time

```python
# Group by date
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'])
accidents_over_time = df.groupby('CRASH DATE').size()

# Plot
plt.figure(figsize=(12, 5))
accidents_over_time.plot(title="Accidents Over Time", color='blue')
plt.xlabel("Date")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
df['YEAR'] = df['CRASH DATE'].dt.year
print(df.groupby('YEAR').size())
```

### Common Contributing Factors

```python
# Get top 10 contributing factors
top_factors = df['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=top_factors.index, y=top_factors.values, palette="magma")
plt.xlabel("Contributing Factor")
plt.ylabel("Count")
plt.title("Top 10 Contributing Factors of Accidents")
plt.xticks(rotation=90)
plt.show()
```

### Heatmap of Correlation

```python
# Convert numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
